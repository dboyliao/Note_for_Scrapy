# Start a Project

## Scrapy Architecture

![scrapy_architecture](http://doc.scrapy.org/en/latest/_images/scrapy_architecture.png)

As you can see, there are many components in scrapy which are allowed to be costumized.
The basic rules you should remember is as following:

1. In your parser, you can either raise (or return) a item or a request.
2. If you raise a request, the scheduler will pass it to downloader and the response it get will be passed to the callback function which you assign when you raise that request.
3. If you raise a item, the scheduler will pass it to the itempiplines and process it.

See following links for detail:
- [Items](http://doc.scrapy.org/en/latest/topics/items.html)
- [ItemPipline](http://doc.scrapy.org/en/latest/topics/item-pipeline.html)
- [Request](http://doc.scrapy.org/en/latest/topics/request-response.html#scrapy.http.Response)
- [Docs](http://doc.scrapy.org/en/latest/intro/overview.html)


## First Step
Use `scrapy startproject [project_name]` to initialize a scrapy project.

Use `scrapy genspider mydomain mydomain.com` and scrapy will generate the default. 

For example, if you type `scrapy startproject dboy` in your terminal, you should see:

![project_architecture](scrapy_project.png)

Type `scrapy genspider book` in the `spiders` directory, you should see:

![genspider](genspider.png)

That is, `scrapy` will generate a spider template for you!

Here we briefly introduce the functionality of each file in your scrapy project:

1. item.py: 
 * Define your item classes in this file. It will control the fields of the items scraped by your spiders. <br> (That is, the data schema)
2. piplines.py
 * This file contains all the itempiplines your spiders are going to use.
3. settings.py
 * settings of your project. See: http://doc.scrapy.org/en/latest/topics/settings.html.
4. spiders:
 * It is a directory where you put your spiders in.
5. scrapy.cfg:
 * Configuration for scrapy including where the setting file is, where to deploy project and project name etc.
 
 
Note: 
1. if lxml fucks you up on Mac OS X, check out [this](http://stackoverflow.com/questions/19548011/cannot-install-lxml-on-mac-os-x-10-9).
2. if there is "egg_info" error, remove setuptools, reinstall it and upgrade it. should solve the issue.


# Notes for Setting

## Prededence

Although you can set options of scrapy through various ways, they have its own precedence.
Here is the list of them in decreasing order of precedence:

- Command line options (most precedence)
- Project settings module
- Default settings per-command
- Default global settings (less precedence)

# Useful Binaries

## 1. scrapy shell [url]
*example*: scrapy shell http://books.com.tw

You will enter a ipython shell with bunch of objects build by scrapy. 
The following are (I think) the most useful ones:

- response: a scrapy Response object which represent the response your spider will receive.
- fetch(url): renew all the build objects such as response with a new url.

Here is how it looks like when you start a scrapy shell:
![scrapy_shell](scrapy_shell.png)

In the shell, you can use `response.xpath()` or `response.css()` to parse your webpage.

They will return a list of `Selector` objects and the you can use the `extract` method to extract out the data.

## 2. scrapy crawl [spider_name]
*example*: scrapy crawl book

With this command, scrapy will start your spider and crawl the webpage specified in your spider for you. Here is how it looks like when the crawler start:

![scrapy_crawl](scrapy_crawl.png)

Once you have all your spiders done, you are ready to deploy yor scrapy project!

# Deploy

Deploying your project into a Scrapyd server typically involves two steps:

- Building a Python egg of your project. This is called “eggifying” your project. You’ll need to install setuptools for this. See Egg caveats below.
- Uploading the egg to the Scrapyd server

Fortunately, you can get above two things done with one simple command: `scrapy deploy`

1. Run `scrapyd` to start a scrapy daemon.
2. Run `scrapy deploy`, you will see some directories and egg files are generated. These files are generated by scrapy automatically according to the settings.

## Deploy Settings in scrapy.cfg
You can specify the target and project you want when you deploy your spiders.
Here is one example:
<div class="highlight">
<pre>
[deploy:scrapyd2]
url = http://scrapyd.mydomain.com/api/scrapyd/
username = dboyliao
password = secret
project = dboy
</pre>
</div>

After deploy, you can list all scrapyd running with command `scrapyd-deploy -l`
<div>
<pre>
scrapyd2                 http://scrapyd.mydomain.com/api/scrapyd/
</pre>
</div>
